```r
#Normalização das Variáveis
install.packages("moments")
library("moments") 
```

```r
Defeito = scale(data1_Fr$numero_defeitos, center = T)
Mcdc = scale(data1_Fr$cobertura_mcdc, center = T)
Condicao = scale(data1_Fr$cobertura_condicao, center = T)
Decisao = scale(data1_Fr$cobertura_decisao, center = T)
hist(Defeito)
hist(Mcdc)
hist(Condicao)
hist(Decisao)
skewness(Defeito)
kurtosis(Defeito)
shapiro.test(Defeito)
skewness(Mcdc)
kurtosis(Mcdc)
shapiro.test(Mcdc)
skewness(Condicao)
kurtosis(Condicao)
shapiro.test(Condicao)
skewness(Decisao)
kurtosis(Decisao)
shapiro.test(Decisao)
```

```r
f_log <- function(x){
  k <- log10(abs(max(x)))
  return(x/(10^k))
}

Defeito = f_log(data1_Fr$numero_defeitos)
Mcdc = f_log(data1_Fr$cobertura_mcdc)
Condicao = f_log(data1_Fr$cobertura_condicao)
Decisao = f_log(data1_Fr$cobertura_decisao)
hist(Defeito)
hist(Mcdc)
hist(Condicao)
hist(Decisao)
skewness(Defeito)
kurtosis(Defeito)
shapiro.test(Defeito)
skewness(Mcdc)
kurtosis(Mcdc)
shapiro.test(Mcdc)
skewness(Condicao)
kurtosis(Condicao)
shapiro.test(Condicao)
skewness(Decisao)
kurtosis(Decisao)
shapiro.test(Decisao)
```

```r
Defeito = sqrt(data1_Fr$numero_defeitos)
Mcdc = sqrt(data1_Fr$cobertura_mcdc)
Condicao = sqrt(data1_Fr$cobertura_condicao)
Decisao = sqrt(data1_Fr$cobertura_decisao)
hist(Defeito)
hist(Mcdc)
hist(Condicao)
hist(Decisao)
skewness(Defeito)
kurtosis(Defeito)
shapiro.test(Defeito)
skewness(Mcdc)
kurtosis(Mcdc)
shapiro.test(Mcdc)
skewness(Condicao)
kurtosis(Condicao)
shapiro.test(Condicao)
skewness(Decisao)
kurtosis(Decisao)
shapiro.test(Decisao)
```

```r
inversaN <- function(x){
  return((-1) / x)
}

Defeito = inversaN(data1_Fr$numero_defeitos)
Mcdc = inversaN(data1_Fr$cobertura_mcdc)
Condicao = inversaN(data1_Fr$cobertura_condicao)
Decisao = inversaN(data1_Fr$cobertura_decisao)
hist(Defeito)
hist(Mcdc)
hist(Condicao)
hist(Decisao)
skewness(Defeito)
kurtosis(Defeito)
shapiro.test(Defeito)
skewness(Mcdc)
kurtosis(Mcdc)
shapiro.test(Mcdc)
skewness(Condicao)
kurtosis(Condicao)
shapiro.test(Condicao)
skewness(Decisao)
kurtosis(Decisao)
shapiro.test(Decisao)
```

```r
cubica <- function(x){
  return(x^(1/3))
}

Defeito = cubica(data1_Fr$numero_defeitos)
Mcdc = cubica(data1_Fr$cobertura_mcdc)
Condicao = cubica(data1_Fr$cobertura_condicao)
Decisao = cubica(data1_Fr$cobertura_decisao)
hist(Defeito)
hist(Mcdc)
hist(Condicao)
hist(Decisao)
skewness(Defeito)
kurtosis(Defeito)
shapiro.test(Defeito)
skewness(Mcdc)
kurtosis(Mcdc)
shapiro.test(Mcdc)
skewness(Condicao)
kurtosis(Condicao)
shapiro.test(Condicao)
skewness(Decisao)
kurtosis(Decisao)
shapiro.test(Decisao)
```
